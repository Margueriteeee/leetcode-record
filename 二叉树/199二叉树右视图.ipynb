# 199 二叉树右视图
你站在二叉树的右侧，按照从顶到底的顺序，返回右侧能看到的节点值

## 思路
只返回二叉树最右侧的元素  

在层序遍历中，每一层遍历到最后的时候再放进result中即可  
- 判断是否到达最后面的元素  
- 当前层的队列为空

```python
import collections
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def rightSideView(self, root:TreeNode) -> list[int]:
        if not root:
            return []
        
        queue = collections.deque([root])
        right_view = []

        while queue:
            level_size = len(queue)
            for i in range(level_size):
                # 当前层队列长度控制循环次数
                # 每次pop就从队列中移出了
                node = queue.popleft()

                if i == level_size - 1:   
                    right_view.append(node.val)
                
                # 先判断节点存在再入队
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
        
        return right_view
```
