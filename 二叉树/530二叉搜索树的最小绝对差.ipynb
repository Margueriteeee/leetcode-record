# 530.二叉搜索树的最小绝对差
输入：所有节点为非负值的二叉搜索树  
输出：树中任意两个节点差的绝对值的最小值  
ps：树中至少有两个节点  

## 思路
二叉搜索树是有序的！   
中序遍历是一个递增的有序数组，求最小差值  

**在二叉搜索树上求什么最值，求差值之类的，都要思考一下二叉搜索树可是有序的，要利用好这一特点**

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None) -> None:
        self.val = val
        self.left = left
        self.right = right

# 递归法，利用中序递增结合数组

class Solution:
    def __init__(self):
        self.vec = []
    
    def traversal(self, root):
        if root is None:
            return
        
        self.traversal(root.left)
        self.vec.append(root.val)
        self.traversal(root.right)
    
    def getMinimumDifference(self,root):
        self.vec = []
        self.traversal(root)
        if lef(self.vec) < 2:
            return 0
        result = float('-inf')
        for i in range(1, len(self.vec)):
            # 统计有序数组的最小差值
            result = min(result, self.vec[i] - self.vec[i - 1])    # 递增的中序数组直接 后减前，就是正的
        return result
```

```python
# 递归优化版：中序遍历本来就是递增的
# 一遍遍历一遍记录前一个结点值作差即可

class Solution:
    def __init__(self):   # 定义全局变量，前一个节点值和求出的最小差值
        self.result = float('-inf')
        self.pre = None
    
    def traversal(self, cur):
        if cur is None:
            return 
        self.traversal(cur.left)  # 左
        if self.pre is not None:  # 中
            self.result = min(self.result, cur.val - self.pre.val)
        self.pre = cur   # 更新前一个结点
        self.traversal(cur.right) # 右

    def getMinimumDifference(self, root):
        self.traversal(root)
        return self.result
```
