# 515 在每个树行中找最大值
输入：二叉树  
输出：二叉树每一层的最大值组成的数组  

## 思路
层序遍历就完事了  

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None) -> None:
        self.val = val
        self.left = left
        self.right = right

import collections

class Solution:
    def largestValues(self, root:TreeNode) -> list[int]:
        if not root:
            return []
        
        largest = []
        queue = collections.deque([root])

        while queue:
            level_size = len(queue)
            max = float('-inf')

            for _ in range(level_size):
                node = queue.popleft()
                max = max(node.val, max)

                if node.left:
                    queue.append(node.left)

                if node.right:
                    queue.append(node.right)
            
            largest.append(max)

        return largest

```
