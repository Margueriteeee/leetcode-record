# 二叉树的层序遍历
层序遍历：逐层地，从左到右访问所有节点



## 思路
借助 队列 实现：先进先出 —— 广度优先遍历

重要的模板，可以打好多题呜呜呜呜

```python
from typing import Optional, List

class TreeNode:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# 法1：利用长度
# 先把根节点加入队列
# 每一层为一个循环，依次把队头元素加入对应层号的列表中，然后把队头元素的两个子节点加入队列
# 在循环的最后把当前层的列表append进result中
# 推荐推荐

class Soluion:
    def levelOrder(self, root: Optional[TreeNode]) -> list[list[int]]:
        # 边界条件：空树返回空列表
        if not root:
            return []
        # 创建一个双端队列作为队列
        queue = collections.deque([root])
        result = []   # 存储按层划分的结果，每层另外是一个list
        while queue:
            level = []
            for _ in range(len(queue)):   # 获取当前层的节点数（每轮循环会变化
                cur = queue.popleft()  # 队头元素出队加入当前层列表
                level.append(cur.val)
                if cur.left:     # 把子节点（下一层）加入队列
                    queue.append(cur.left)
                if cur.right:
                    queue.append(cur.right)
            result.append(level)  # 当前层的结果加入最终列表
        
        return result
```

# 层次遍历Ⅱ
给定二叉树，返回节点自底向上的层次遍历  
逐层从左向右遍历

## 思路
上面的层序遍历结果数组反转一下就可以了

```python
class Solution:
    def levelOrderBottom(self, root:TreeNode) -> list[list[int]]:
        if not root:
            return []
        
        result = []
        queue = collections.deque([root])

        while queue:
            level = []
            for _ in range(len(queue)):
                cur = queue.popleft()
                level.append(cur.val)
                if cur.left:
                    queue.append(cur.left)
                if cur.right:
                    queue.append(cur.right)
            result.append(level)
        
        return result[::-1]  # 最后输出的数组翻转一下
```
