# 513 找树左下角的值
给定一个二叉树，返回其最后一行最左边的节点值



## 思路
首先要在最后一层去找，也就是必须是一个叶子节点（左右孩子均为空）  

最后一层的第一个元素即为目标值

最直观的就是迭代法！也最简单

```python
# 层序遍历法
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

from collections import deque
class Solution:
    def findBottomLeftLeaves(self, root:TreeNode) -> int:
        if not root:
            return 0
        
        queue = deque()
        queue.append(root)
        result = 0

        while queue:
            size = len(queue)
            for i in range(size):
                node = queue.popleft()
                if i == 0:   # 每一层的第一个值都记录下来，不断覆盖直到得到最后一层的第一个值即为结果
                    result = node.val
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(left.right)
        return result
```
