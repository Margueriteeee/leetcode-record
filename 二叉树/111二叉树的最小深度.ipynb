# 111.二叉树的最小深度
二叉树的最小深度 ： 根节点到**最近的叶子节点的最短路径**上的节点数量

区分一下求深度和求高度的区别：

- 节点的深度：  
  - 从根节点开始，到该节点的最长简单路径的边的条数或者节点数（取决于深度从0开始还是从1开始） 
  - 前序遍历 
- 节点的高度：从该节点开始，到叶子节点的最长简单路径的边的条数或者节点数（取决于深度从0开始还是从1开始）
  - 后序遍历  

这题使用层序遍历进行迭代也是可以的，写法在 104二叉树的最大深度 这篇中有

### 递归法
递归三部曲上来！

1. 递归函数的参数和返回值  
   - 参数：传入的二叉树的根节点
   - 返回值：int表示深度

2. 终止条件  
   - 遇到空节点就需要返回了
   - 孔节点返回0，表示当前节点的高度是0（叶子节点）

3. 递归逻辑
   - 对于一个节点，如果左子树非空，右子树为空，索命最小深度是 1+左子树的深度  
   - 同样对于左子树为空，右子树非空，则最小深度为 1+右子树深度
   - 左右子树都不为空，则取两个子树的最小深度 +1

```python
# 递归法
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def getDepth(self, node):
        if not node:
            return 0
        
        leftDepth = self.getDepth(node.left)
        rightDepth = self.getDepth(node.right)
        # 左不空右空
        if node.left is not None and node.right is None:
            return 1 + leftDepth
        if node.right is not None and node.left is None:
            return 1 + rightDepth
        
        result = 1+ min(rightDepth, leftDepth)
        return result

    def minDepth(self, root:TreeNode) -> int:
        return self.getDepth(root)
```

```python
# 层序遍历迭代法，复习一下
import collections
class Solution:
    def minDepth(self, root:TreeNode) -> int:
        if not root:
            return 0
        
        depth = 0
        queue = collections.deque([root])

        while queue:
            # 进入循环就有下一层，层数+1
            depth += 1

            for _ in range(len(queue)):   # 遍历当前层
                node = queue.popleft()
                if not node.left and not node.right:
                    return depth

                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
        
        return depth
```
