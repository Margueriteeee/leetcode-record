# 701.二叉搜索树中的插入操作
输入：给定二叉搜索树，以及要插入的节点的值  
输出：把节点值插入二叉搜索树后的根节点  

ps：插入的值和树中所有值都不同  
ps：可以使用任意有效的插入方式，插入后仍保持二叉搜索树即可  

## 思路
需要插入的本来就是一个有序的二叉搜索树，只需要找到符合条件的空节点插入即可  

- 遍历二叉搜索树，找到空节点插入元素  
- 根本不用重构二叉搜索树！

### 递归
1. 递归参数和返回值
   - 参数：根节点，需要插入的值  
   - 返回值：根据返回值完成新加入的节点和父节点的赋值操作

2. 确定终止条件  
   - 递归到空姐点就是需要插入节点的位置了  
   - 返回需要插入的节点

3. 单层递归逻辑
   - 也不需要遍历整棵树  
   - 根据插入值的大小比较搜索的左右方向就可以了

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None) -> None:
        self.val = val
        self.left = left
        self.right = right

# 递归（无返回值，递归中找到空节点直接加入）
        
class Solution:
    def __init__(self):
        self.parent = None

    def traversal(self, cur, val):
        if cur is None:   # 终止条件:找到空节点
            node = TreeNode(val)
            if val > self.parent.val:
                self.parent.right = node
            else:
                self.parent.left = node
            return
        
        self.parent = cur  # 单层递归逻辑
        if cur.val > val:
            self.traversal(cur.left, val)
        if cur.val < val:
            self.traversal(cur.right, val)
    
    def insertIntoBST(self, root, val):
        self.parent = TreeNode(0)
        if root is None:
            return TreeNode(val)
        self.traversal(root, val)
        return root
```

```python
# 递归二，有返回值

class Solution:
    def insertIntoBST(self, root, val):
        if root is None or root.val == val:
            return TreeNode(val)
        
        elif root.val > val:
            if root.left is None:   # 左节点为空，直接加入
                root.left = TreeNode(val)
            else:                   # 左节点不为空，继续递归直到找到左节点有空位
                self.insertIntoBST(root.left, val)
        
        elif root.val < val:
            if root.right is not None:
                root.right = TreeNode(val) # 右节点为空，直接创建新节点加入
            else:
                self.insertIntoBST(root.right, val)  # 右节点无空位，继续向右找更大的空位

        return root
```

```python
# 迭代法
# 好理解呜呜呜

class Solution:
    def insertIntoBST(self, root, val):
        if root is None: # 如果根节点为空，创建新节点作为根节点返回  
            return TreeNode(val)
        
        cur = root
        parent = root  # 记录上一个节点（新节点的父节点）
        while cur is not None:
            parent = cur    # 更新上一个节点
            if cur.val > val:   # 更新当前节点
                cur = cur.left
            else:
                cur = cur.right
                
        if val < parent.val:
            parent.left = TreeNode(val)   # 新节点插入父节点的左子树
        else:
            parent.right = TreeNode(val)  # 新节点连接到父节点的右子树

        return root
```
