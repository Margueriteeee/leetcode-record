```python
class Node:
    def __init__(self, val: int = 0, left: 'Node' = None, right: 'Node' = None, next: 'Node' = None):
        self.val = val
        self.left = left
        self.right = right
        self.next = next

import collections

class Solution:
    def connect(self, root:'Node') -> 'Node':
        if not root:
            return root

        queue = collections.deque([root])

        while queue:
            level_size = len(queue)
            prev = None   # 

            for _ in range(level_size):
                node = queue.popleft()  # 

                if prev:
                    prev.next = node   # prevnext
                
                prev = node

                if node.left:
                    queue.append(node.left)
                
                if node.right:
                    queue.append(node.right)

        return root
```
