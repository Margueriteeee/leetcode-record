# 501 二叉搜索树中的众数
输入：有相同值的二叉搜索树  
输出：所有出现频率最高的元素  

## 非二叉搜索树的做法
普通的二叉树是无序的，直观做法就是遍历，字典或集合统计频率  

### 递归
1. 遍历整个树（随便遍历方式），统计所有元素的频率  
2. 把统计出的频率进行排序  
3. 输出前面频率最高的元素

## 二叉搜索树的做法
还是利用二叉搜索树中序遍历为递增数组的规律  

1. 中序遍历得到一个递增数组 

2. 直接在中序遍历的过程中统计频率：  
   - 技巧：**pre指针 和 cur指针** 
   -  pre指针指向前一个结点，每次把当前节点和前一个结点进行比较  
   -  **只遍历一遍**：如果 当前频率==最大频率，吧当前元素加入结果集合中  
   -  为了防止后续出现更大频率的元素，每次更新结果集合时需要把之前的元素清空（确保当前放进去的是最大的）  

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None) -> None:
        self.val = val
        self.left = left
        self.right = right

# 递归法，使用字典

from collections import defaultdict
class Solution:
    def searchBST(self, cur, freq_map):
        if cur is None:
            return
        freq_map[cur.val] += 1   # 统计元素频率
        self.searchBST(cur.left, freq_map)
        self.searchBST(cur.right, freq_map)

    def findMode(self, root):
        freq_map = defaultdict(int)    # key:元素，value:出现频率
        result = []
        if root is None:
            return result
        self.searchBST(root, freq_map)
        max_freq = max(freq_map.values())    # values转换为列表取max
        for key, freq in freq_map.items():
            if freq == max_freq:
                result.append(key)
        return result
```

```python
# 迭代法,栈模拟
# 双指针技巧，只遍历一次

class Solution:
    def findMode(self, root):
        st = []
        cur = root
        pre = None
        maxCount = 0   # 最大频率
        count = 0      # 统计频率
        result = []

        while cur is not None or st:
            if cur is not None:    # 指针来访问节点，先从左节点访问到最底层
                st.append(cur)
                cur = cur.left
            else:
                cur = st.pop()
                if pre is None:  # 第一个节点
                    count = 1
                elif pre.val == cur.val:  # 当前节点与前一个节点相同，统计频率+1
                    count += 1
                else:     # 与前一个结点不同
                    count = 1

                if count == maxCount:   # 和最大频率相同，计入结果数组
                    result.append(cur.val)

                if count > maxCount:
                    maxCount = count  # 更新最大频率
                    result = [cur.val]  # 清空结果数组，只保留当前值
                
                pre = cur        # 更新指针
                cur = cur.right  # 右
        
        return result
```
