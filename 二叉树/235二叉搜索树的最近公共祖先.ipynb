# 235. 二叉搜索树的最近公共祖先
输入：给定的二叉搜索树，和两个指定节点  
输出：该 二叉搜索树 两个指定节点的最近公共祖先  

## 思路
本题和236不同，本题规定了给定的树是有序的二叉搜索树  

但总体思路仍然是：判断一个结点的左子树有p，右子树有q  

利用好二叉搜索树：**中序遍历递增的性质**

- 中序遍历数组中，中节点一定在区间[p,q]里面  
- 从根节点开始向下遍历，**第一个**遇到的位于区间内的节点，即为最近的公共祖先  

而且不需要遍历整棵树，遍历到了直接返回就行

### 递归三部曲
1. 参数和返回值  
   - 参数：当前节点和两个值p, q  
   - 返回值：最近的公共祖先

2. 终止条件
   - 递归到空节点就返回空
   - 其实也不是需要这个终止条件，因为q，p是一定存在的
   - 所以肯定可以找到符合条件的公共祖先

3. 单层递归逻辑  
   - 遍历的范围是区间 [p, q] (不知道p，q谁大，两个都要比较)
   - 每遍历到一个节点，先确定val是否在这个区间里  
      - val < p && val < q: 说明小了，应该向右遍历  
      - val > p && val > q: 说明大了，应该向左遍历
   - 遇到val在区间里面，直接就返回

对于有序二叉搜索树的操作不用回溯，利用好有序性

```python
class TreeNode:
    def __init__(self, val=0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def traversal(self, cur, p, q):
        if cur is None:
            return cur    # 可以省略，不会遍历到空结点
        
        if cur.val > p.val and cur.val > q.val:
            left = self.traversal(cur.left, p, q)
            if left is not None:
                return left
        
        if cur.val < p.val and cur.val < q.val:
            right = self.traversal(cur.right, p, q)
            if right is not None:
                return right
            
        return cur
    
    def lowestCommonAncestor(self, root, p, q):
        return self.traversal(root, p, q)
```

```python
# 迭代法
# 更简单！！
class Solution:
    def lowestCommonAncestor(self, root, p, q):
        while root:
            if root.val > p.val and root.val > q.val:
                root = root.left
            elif root.val < p.val and root.val < q.val:
                root = root.right
            else:   # 刚好处在区间内，直接返回当前节点
                return root
            
        return None
```
