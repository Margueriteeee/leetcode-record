# 二叉树的递归遍历

## 递归三要素
只要写递归，就找齐下面的三要素：  

1. 递归函数的参数和返回值：
   - 递归过程中需要处理哪些参数，在函数里把它加上    
   - 明确每次递归返回的是什么值，确定递归函数的返回类型  
  
2. 确定终止条件
   - 如果运行递归出现了 *栈溢出* ，说明终止条件没写 / 写的不对  
   - 递归的每一层是用栈来保存的  
   - 递归不终止，必然会溢出  
  
3. 确定单层递归的逻辑
   - 每一层递归中需要处理的信息  

### eg：二叉树的深度优先搜索 DFS

**前序遍历**

1. 递归函数的参数和返回值：  
   - 参数：当前要处理的二叉树节点 node，  
   - 返回值: None, 因为每次递归遍历的结果都存放在 res 中了  

2. 终止条件
   - 如果当前节点为空，说明已经处理完所有的节点了，此时可以终止（递归的出口）  
   - node == None
  
3. 单层递归逻辑
   - 前序遍历：父节点 -> 左孩子 -> 右孩子   
   - 访问当前节点，把当前节点的val加入res  
   - 递归处理其左节点
   - 递归处理其右节点

```python
class TreeNode:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# LC144 二叉树的前序遍历
class Soluion:
    def preorderTraversal(self, root:TreeNode) -> list[int]:
        res = []

        def dfs(node):
            if node is None:   # 终止条件
                return     # 返回值为空

            res.append(node.val)   # 递归逻辑:中左右
            dfs(node.left)
            dfs(node.right)
        
        dfs(root)

        return res
```

```python
# LC94 二叉树的中序遍历
class Soluion:
    def inorderTraversal(self, root:TreeNode) -> list[int]:
        res = []
        
        def dfs(node):
            if node is None:
                return 

            dfs(node.left)       # 递归逻辑：左中右 
            res.append(node.val)
            dfs(node.right)

        dfs(root)

        return res
```

```python
# LC145 二叉树的后序遍历
class Soluion:
    def postorderTraversal(self, root:TreeNode) -> list[int]:
        res = []

        def dfs(node):
            if node is None:
                return
            
            dfs(node.left)
            dfs(node.right)
            res.append(node.val)

        dfs(root)

        return res
```
