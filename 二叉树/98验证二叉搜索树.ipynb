# 98.验证二叉搜索树
判断给定的二叉搜索树，是否是一个有效的二叉搜索树  

有效的二叉搜索树：  
- 节点的左子树只包含小于当前节点的数 -> 左子树所有节点**都**小于当前节点  
- 节点的右子树只包含大于当前节点的数 -> 右子树所有节点**都**大于当前节点
- 所有左右子树都是二叉搜索树  

## 思路  
**中序遍历下，二叉搜索树的数值是一个有序序列（增）**

所以，把输入的二叉树通过中序遍历，验证其中序遍历的序列是否递增即可  

注意：根据要求，右子树中不可以有小于当前节点的值，不能单纯的跟子树的根节点比较  


```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None) -> None:
        self.val = val
        self.left = left
        self.right = right

# 递归法，设定极小值进行比较
# 寻找的是不符合条件的节点，因此需要返回值
# 当二叉树为空时也是一个有效的二叉搜索树
# 单层递归中按照左中右的中序遍历顺序，遇到不符合条件的节点就返回，否则遍历整个树

class Solution:
    def __init__(self):
        self.maxVal = float('-inf')  # 因为后台测试数据中有int最小值

    def isValidBST(self, root):
        if root is None:
            return True
        
        left = self.isValidBST(root.left)   # 左
        # 中序遍历，验证遍历的元素是否从小到大
        if self.maxVal < root.val:   # 中
            self.maxVal = root.val
        else:
            return False    # 出现值比记录还要小说明不是有效的
        
        right = self.isValidBST(root.right)  # 右

        return left and right    # 左右子树均为有效二叉搜索树才返回true
```
