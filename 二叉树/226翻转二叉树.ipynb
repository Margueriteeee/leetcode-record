# 226 翻转二叉树
把一颗二叉树每一层从左到右翻转



## 思路
整个树以中线分割线进行翻转，类似于‘轴对称’

落到每一个节点就是把其左右孩子交换位置  
- 改变指针
- 前序和后序遍历比中序方便得多
- 中序会把其中一些节点反转两次

**二叉树的问题一定要想清楚用的是什么遍历方式**  

记得住迭代最好先用迭代，递归在工程中需要谨慎使用，  

面试中也可能在写出递归的基础上进一步考察迭代

```python
# 首选：前序迭代法
class TreeNode:
    def __init__(self, val=0, left = None, right=None) -> None:
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def invertTree(self, root:TreeNode) -> TreeNode:
        if not root:
            return None
        
        stack = [root]
        while stack:
            # 处理中节点
            node = stack.pop()
            node.left, node.right = node.right, node.left
            if node.right:
                stack.append(node.right)
            if node.left:
                stack.append(node.left)

        return root
```

```python
# 迭代法层序遍历
import collections
class Solution:
    def invertTree(self, root:TreeNode) -> TreeNode:
        if not root:
            return None
        
        queue = collections.deque([root])

        while queue:
            node = queue.popleft()
            node.left, node.right = node.right, node.left
            if node.left: queue.append(node.left)
            if node.right:queue.append(node.right)

        return root
```

```python
# 前序递归法

class Solution:
    def invertTree(self, root: TreeNode) -> TreeNode:
        if not root:
            return None
        root.left, root.right = root.right, root.left
        self.invertTree(root.left)
        self.invertTree(root.right)
        return root
```
