# 700.二叉搜索树中的搜索
输入：二叉搜索树和指定的搜索值   
输出：以搜索值为根节点的子树

  

## 思路
二叉搜索树：节点的值是经过排序的  
- 左子树不空，则左子树上所有节点都小于根节点  
- 右子树不空，则右子树上所有结点的值均大于根节点值
- 左右子树均为二叉搜索树  

> 
> **二叉搜索树的遍历方式和普通二叉树是不同的**  
>

### 递归遍历  
1. 递归参数和返回值  
   - 递归参数：根节点、搜索的目标值  
   - 返回值：搜索数值所在的节点

2. 终止条件  
   - 传入的root为空、或者找到目标值，即返回root节点

3. 单层递归逻辑  
   - 二叉搜索树的节点是有序的，所以搜索也可以是有序的  
   - 如果root的值 > target，就搜索左子树（小于根节点）  
   - 如果root的值 < target, 搜索右子树（大于根节点）  


```python
class TreeNode:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# 递归遍历
        # 为什么要有返回值: 
        #   因为搜索到目标节点就要立即return，
        #   这样才是找到节点就返回（搜索某一条边），如果不加return，就是遍历整棵树了。
class Solution:
    def searchBST(self, root:TreeNode, target: int) -> TreeNode:
        if not root or root.val == target:
            return root
        
        if root.val > target:
            return self.searchBST(root.left, target)
        
        if root.val < target:
            return self.searchBST(root.right, target)
```

### 迭代法
普通的二叉树采用迭代法需要用栈或者队列来模拟  

但二叉搜索树是有序的，所以不需要回溯，因为节点的有序性就给出了搜索的方向  

判断根节点和目标值的大小 就确定了往左还是往右搜索

```python
# 迭代法
class Solution:
    def searchBST(self, root:TreeNode, target:int) -> TreeNode:
        while root:
            if root.val > target: root = root.left
            elif root.val < target: root = root.right
            else: return root

        return None   # 遍历到这说明没找到
```
