# 236. 二叉树的最近公共祖先
输入：给定的二叉树，两个指定节点  
输出：两个指定节点的最近公共祖先  

最近公共祖先的定义：  
- 有根树 T 的两个结点 p、q  
- 最近公共祖先x满足 x 是 p、q 的祖先且 x 的深度尽可能大（一个节点也可以是它自己的祖先）

ps：所有节点值都是唯一的（互不重复）

## 思路
如果可以**自底向上**查找，就很方便找到公共祖先了  

### 回溯
回溯是典型的从底向上的过程 

对应的**后序遍历**：左右中，根据左右子树的返回值来处理中间节点  

判断一个节点是指定节点p、q的公共祖先：  
- 左子树包含p且右子树包含q  
- 或者，左子树包含q， 右子树包含p  

使用递归法：  
- 遍历到p把p返回，遍历到q把q返回（p、q一定存在）    
- 如果一个节点的左右子树返回值都不为空，则该节点就是最近祖先  

特殊情况：p或q本身就是二者的公共祖先，这包含在上述的递归逻辑中  

#### 递归三部曲
1. 递归参数和返回值：  
   - 参数：根节点  
   - 返回值：节点，反映是否找到节点p或者q  

2. 终止条件：  
   - 遇到空节点返回空  

3. 单层递归逻辑：  
   - 根据返回值进行判断  
   - 用两个变量接住左右子树的递归返回值  
   - 但其实在所有子树的递归过程中已经遍历完整棵树了  
   - 左右子树都不为空，返回当前节点值（最近公共祖先）  
   - 左空右不空，返回右，右空左不空返回左  
   - 左右都空，返回空

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None) -> None:
        self.val = val
        self.left = left
        self.right = right

# 递归法，回溯
# 迭代法不适合回溯  
        
class Solution:
    def lowestCommonAncestor(self, root, p, q):
        if root == q or root == p or root is None:
            return root
        
        left = self.lowestCommonAncestor(root.left, p, q)
        right = self.lowestCommonAncestor(root.right, p, q)

        if left is not None and right is not None:
            return root
        if left is None and right is not None:
            return right
        elif left is not None and right is None:
            return left
        
        else:
            return None
```
