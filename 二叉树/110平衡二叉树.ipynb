# 110 平衡二叉树
判断一个二叉树是否为平衡二叉树

平衡二叉树：每个节点 左右两个子树的高度差 绝对值不超过1

## 思路
leetcode定义：根节点的深度为1  

在 二叉树的最大深度 这一题中，求最大的深度其实等同于求 **根节点的高度**  
- 所以使用的是 后序遍历
- 因为 根节点的高度= 左右子树的高度 + 1  

所以其实**求高度**用的是 **后序遍历**  

求**深度**真正的逻辑是 **前序遍历** ！  
- 中左右的节点访问顺序，先判断当前的中间节点是否达到最大深度

所以本题中要比较的是左右两个子树的高度差，也要用到 后序遍历

### 递归：
1. 参数和返回值  
   - 参数：当前传入的节点  
   - 返回值：当前节点为根节点的树的高度  
 - 标记左右子树的高度差？  
 - 如果当前节点为根节点的树已经不是平衡二叉树了，返回false

2. 终止条件：当前节点为空，返回0  

3. 递归逻辑：判断当前树是否为平衡二叉树  
   - 分别求出其左右子树的高度  
   - 比较 差值和1 的大小，大于1返回false
   - 小于1 返回true

Ps：  
求最大深度的时候可以使用层序遍历，因为求深度和层序遍历都是自顶向下的过程  

但本题求高度的过程如果使用层序遍历则效率很低

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def isBalanced(self, root:TreeNode) -> bool:
        if self.get_height(root) != -1:
            return True
        else:
            return False

    def get_height(self, root:TreeNode) -> int:
        if not root:
            return 0
        
        # 计算左子树高度
        if (left_height := self.get_height(root.left)) == -1:
            return -1
        
        # 计算右子树高度
        # 海象运算符：= 等价于：
        # right_height = self.get_height(root.right)
        # if right_height == -1:
        #     return -1
        if (right_height := self.get_height(root.right)) == -1:
            return -1   # 如果子树不平衡直接返回-1
        
        # 递归处理逻辑，判断是否平衡
        if abs(left_height - right_height) > 1:
            return -1
        else:
            return 1 + max(left_height, right_height)   # 子树平衡的情况返回根节点的高度
```
