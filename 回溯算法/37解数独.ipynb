# 37 解数独
编写程序，填充数独问题中的空格

# 思路
棋盘中的每一个位置都要放一个数字，并检查数字是否合法

合法规则：
- 每个3×3的小棋盘中1-9每个数字只能出现一次  
- 大棋盘中每行每列1-9每个数字只能出现一次

## 回溯三部曲
1. 递归函数及其参数
   -  返回值：bool，表示是否找到对应位置的合法数字

2. 终止条件：无需，遍历所有叶子和整个树形结构，找到合法的立刻返回

3. 单层搜索逻辑：行列方向二维递归  
      -  两个for循环分别遍历行列两个方向
      -  找到符合条件的数就返回
      -  如果9个数都不行则该棋盘无解，直接返回

```python
class Solution:
    def solveSudoku(self, board:list[list[str]]) -> None:
        row_used = [set() for _ in range(9)]
        col_used = [set() for _ in range(9)]
        box_used = [set() for _ in range(9)]
        for row in range(9):
            for col in range(9):
                num = board[row][col]
                if num == '.':
                    continue
                row_used[row].add(num)
                col_used[col].add(num)
                box_used[(row // 3) * 3 + col // 3].add(num)
        self.backtracing(0, 0, board, row_used, col_used, box_used)
    
    def backtracing(
            self,
            row:int,
            col:int,
            board:list[list[str]],
            row_used:list[list[int]],
            col_used:list[list[str]],
            box_used:list[list[int]],
    ) -> bool:
        if row == 9:
            return True
        
        next_row, next_col = (row, col+1) if col<8 else((row+1, 0))
        if board[row][col] != '.':
            return self.backtracing(next_row, next_col, board, row_used, col_used, box_used)
        
        for num in map(str, range(1, 10)):
            if(
                num not in row_used[row]
                and num not in col_used[col]
                and num not in box_used[(row // 3) * 3 + col // 3]
            ):# 当前数字都没有被使用过，开始遍历
                board[row][col] = num
                row_used[row].add(num)
                col_used[col].add(num)
                box_used[(row // 3) * 3 + col // 3].add(num)
                if self.backtracing(
                    next_row, next_col, board, row_used, col_used, box_used
                ):
                    return True  # 回溯递归能找到对应位置返回true
                
                # 回溯撤销
                board[row][col] = "."
                row_used[row].remove(num)
                col_used[col].remove(num)
                box_used[(row // 3) * 3 + col // 3].remove(num)
        
        # 上面遍历都没找到，返回false
        return False
```
