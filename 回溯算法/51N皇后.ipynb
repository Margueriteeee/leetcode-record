```python
class Solution:
    def solveQueens(self, n:int) -> list[list[str]]:
        result = []
        chessboard = ['.' * n for _ in range(n)]  # ,
        self.backtracing(n, 0, chessboard, result)
        # 
        return [[''.join(row) for row in solution] for solution in result]
    
    def backtracing(self, n, row, chessboard, result):
        if row == n:
            result.append(chessboard[:])   # 
            return 
        
        for col in range(n):
            if self.isValid(row, col, chessboard):
                chessboard[row] = chessboard[row][:col] + 'Q' + chessboard
```
