```python
class Solution:
    def permute(self, nums):
        result = []
        self.backtracing(nums, [], [False]*len(nums), result)
        return result
    
    def backtracing(self, nums, path, used, result):
        if len(path) == len(nums):
            result.append(path[:])
            return
        
        for i in range(len(nums)):
            if used[i]:  continue
            used[i] = True  # 
            path.append(nums[i])
            self.backtracing(nums, path, used, result)
            path.pop()
            used[i] = False
```
