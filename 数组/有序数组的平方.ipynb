# 有序数组的平方
给定一个非递减顺序的有序数组，返回每个数字平方后按非递减顺序排列的数组  
- *存在负数，平方后排序变大！*  

## 暴力解法
先平方后排序，快速排序方法

## 双指针法
负数元素平方后可能成为最大数，只在两边不在中间  

用两个指针标记原数组的两端，采用双指针法  
- i：元素的初始位置  
- j：元素的最终位置
- k：指向新数组的终止位置    

```python
# 双指针法
class Solution:
    def sortedSquares(self, nums:list[int]) -> list[int]:
        l, r, i = 0, len(nums)-1, len(nums)-1
        result = [float('inf')] * len(nums)
        while l <= r:
            # 左指针小于右指针
            if nums[l] ** 2 < nums[r] ** 2:
                result[i] = nums[r] ** 2
                r -= 1 # 右边的更大，优先右指针往左移动
            else:
                result[i] = nums[l] ** 2
                l += 1
            i -= 1  # 结果数组中指针左移
        return result

nums = [-4, -2, 4, 6]
solution = Solution()
result = solution.sortedSquares(nums)
result
```

```python
# 暴力排序法
class Solution:
    def sortedSquares(self, nums:list[int]) -> list[int]:
        for i in range(len(nums)):
            nums[i] *= nums[i]
        nums.sort()  # 直接调用排序方法，在原数组内存上排序（无返回值）
        return nums

nums = [-4, -2, 4, 6]
solution = Solution()
result = solution.sortedSquares(nums)
result
```

```python
# 暴力排序+列表推导
class Solution:
    def sortedSquares(self, nums:list[int]) -> list[int]:
        return sorted(x*x for x in nums)
```

## 补充：快速排序算法
核心思想：分区 + 递归  
先选定一个轴 pivot（一般直接选末尾元素）  
- 遍历所有元素，大于pivot放右边，小于放左边    
- 第一次分区后pivot左右都是有序的  
- 再递归对pivot左右的子区间排序
