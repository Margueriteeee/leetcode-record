# 数组篇总结
## 数组基础
- 数组下标从零开始  
- 数组内存空间的地址连续  
- 增删元素的本质是移动其他元素的地址  

## 二分法  
- 写好二分法的核心：坚持 ***循环不变量***  
- 保持区间的定义不变，左闭右开最好记  
  - 暴力解法时间复杂度：O(n)  
  - 二分法时间复杂度：O(logn)  

## 双指针法  
用一快一慢两个指针，在一个for循环下完成两个循环的工作  
- **数组和链表** 中常用  
- 数组在内存中不能释放单一元素，只能移动其他元素实现覆盖  
- 核心要义： ***弄清楚两个指针移动的条件***  、

## 滑动窗口  
- 窗口长度  
- 窗口内容  
- 窗口起始位置  
- 根据当前子序列和大小的情况，不断调整子序列的起始位置，改变窗口大小  
- 从而将O(n^2)的暴力解法降为O(n)

## 模拟行为
- 边界条件较多  
- 坚持 **循环不变量** 原则  
- 弄清楚模拟过程中每次循环的边界条件（左开右闭）
  - 循环次数  
  - 特殊元素/情况  
  - 每次循环的起始位置  

## 前缀和  
把每次查询的过程简化，遍历的过程中求和并记录  
利用**求差**简化查找并求和的过程
