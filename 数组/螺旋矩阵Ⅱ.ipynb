# 螺旋矩阵
输入：一个正整数n，  
输出：正方形矩阵，包含 1~n^2 所有元素，且按顺时针顺序螺旋排列  

*核心：坚持每条边的条件相同*  
统一为左闭右开  
遇到拐角变为新的循环  



```python
class Solution:
    def generateMatrix(self, n:int) -> list[list[int]]:
        nums = [[0] * n for _ in range(n)]  # 创建初始正方形矩阵
        startx, starty = 0, 0          # 起始点，每一圈都会变化
        loop, mid = n // 2, n // 2     # 迭代次数(转的圈数），n为奇数时矩阵的中心点
        count = 1                      # 计数

# python中的range本身就是左闭右开的
        for offset in range(1, loop + 1):   # 每循环一层偏移量+1，offset从1开始(左闭右开)
            for j in range(starty, n - offset):    # 从左至右，左闭右开
                nums[startx][j] = count
                count += 1
            for i in range(startx, n - offset):    # 从上往下
                nums[i][n - offset] = count   # 此时starty已经指向最后一列
                count += 1
            for j in range(n - offset, starty, -1):  # 从右到左，步长-1表示每次减一
                nums[n-offset][j] = count
                count += 1
            for i in range(n-offset, startx, -1):  # 从下到上
                nums[i][starty] = count
                count += 1
            startx += 1
            starty += 1

        if n % 2 != 0:     # n 为奇数时中心点单独标
            nums[mid][mid] = count
        
        return nums

solution = Solution()
n = 5
nums = solution.generateMatrix(n)
nums
```
