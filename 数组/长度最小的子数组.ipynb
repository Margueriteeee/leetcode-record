# 209 长度最小的子数组
输入：一个含n个正整数的数组，一个正整数s  
输出：数组中满足和 >=s 的长度最小的*连续*子数组，及子数组的长度；不存在则返回0  

## 思路
### 暴力解法
两个for循环分别寻找子序列的起始位置和终止位置，不断寻找子序列，时间复杂度n^2  
会超时  

### **滑动窗口法**
滑动窗口：不断调节子序列的起始和终止位置，直到满足相应条件  
其实也算一种双指针法  
- 记录窗口的长度
- 记录窗口内元素之和（内容）
- 标记窗口的起始位置，通过长度得到终止位置

核心：*不断改变起始位置*
- 窗口之和小了：增加窗口长度  
- 窗口之和大了：移动起始位置（减小长度）  

```python
# 暴力解法
class Solution:
    def minSubArratLen(self, s:int, nums:list[int]) -> int:
        l = len(nums)
        min_len = float('inf')  # 传入一个正无穷大的数并转换为浮点数，后面不断替换为找到的最小值

        for i in range(l):
            cur_sum = 0
            for j in range(i, l):
                cur_sum += nums[j]
                if cur_sum >= s:
                    min_len = min(min_len, l-i+1)
                    break
        
        return min_len if min_len != float('inf') else 0
```

```python
# 滑动窗口法
class Solution:
    def minSubArratLen(self, s:int, nums:list[int]) -> int:
        l = len(nums)
        left = 0
        right = 0
        min_len = float('inf')
        cur_sum = 0

        while right < l:
            cur_sum += nums[right]

            while cur_sum >= s:
                min_len = min(min_len, right - left + 1)
                cur_sum -= nums[left]   # 窗口变小减掉当前left的值
                left += 1

            right += 1
        
        return min_len if min_len != float('inf') else 0
```
