# 移除元素  
输入一个数组和要删除的元素，返回删除元素后数组的长度，并把剩下的元素移到原数组的前面  

## 解法一：暴力求解
两层for循环，一个遍历数组元素，一个循环更新数组  
第二个循环中遇到需要删除的值就把后面的所有元素往前移动一格***覆盖***  

## 解法二：双指针法
定义快慢两个指针  
- fast：寻找新数组中的元素
- low：指向更新后新数组的最后一个元素  

删除过程：  
fast逐个遍历，遇到非目标元素：low与fast同步移动  
遇到目标元素：fast++，low不动，并把fast指向的元素覆盖到low上  

```python
import builtins
len = builtins.len
# 双循环暴力求解

class Solution:
    def RemoveElement(self, nums:list[int], val:int) -> int:
        i, l = 0, len(nums)
        while i < l:
            if nums[i] == val:
                for j in range(i+1, l):  
                    # 对i后面的所有元素向前移动一格
                    nums[j-1] = nums[j]
                l -= 1
                i -= 1  # i需要回退一个,避免跳过相邻的val
            i += 1
        return l

solution = Solution()
nums = [1,2,3,4,5,5,7,6,5,9]
val = 2
new_len = solution.RemoveElement(nums, val)
print(new_len, nums)
```

len是关键函数名，不要随便当成变量名！!!  
不然就把原来函数给覆盖了呜呜呜  

```python
# 双指针法
class Solution:
    def RemoveElement(self, nums:list[int], val:int) -> int:
        fast = 0
        low = 0
        size = len(nums)
        while fast < size:  # 遍历所有元素,n个元素实际下标为0~n-1
            # slow用来收集不等于val的值
            if nums[fast] != val:
                nums[low] = nums[fast]
                slow += 1  # 收集完当前元素后+1用以指示数组长度
            fast += 1
        return slow
```
