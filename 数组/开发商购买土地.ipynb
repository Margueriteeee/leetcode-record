# 44 开发商购买土地
输入：n×m的矩阵  
要求：按行或者列的方向分两组，使两组元素和的差异最小  
输出：可能的最小差距  

**基于前缀和**求解：  
将行方向和列方向的和分别求出前缀和  
- 优化：行列分别遍历的时候，到末尾统一求和一下即可  


```python
# 优化版
import sys

def main():
    input = sys.stdin.read
    data = input().split()

    idx = 0
    n = int(data[idx])
    idx += 1
    m = int(data[idx])
    idx += 1  # idx = 2
    sum = 0
    vec = []

    for i in range(n):
        row = []
        for j in range(m):
            num = int(data[idx])   # 按行形成矩阵
            idx += 1
            row.append(num)
            sum += num
        vec.append(row)

    result = float('inf')
    count = 0

    # 行切分
    for i in range(n):
        for j in range(m):
            count += vec[i][j]
            # 遍历到一行的末尾开始统计
            if j == m-1:
                # 如果当前的总和count符合最小的要求，那么两组之和最大也就是所有元素之和sum
                result = min(result, abs(sum - 2*count))
    
    count = 0
    # 列切分
    for j in range(m):
        for i in range(n):
            count += vec[i][j]
            if i == n-1:
                result = min(result, abs(sum - 2*count))
    
    print(result)

if __name__=="__main__":
    main()
```
