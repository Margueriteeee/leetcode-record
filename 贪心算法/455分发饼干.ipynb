# 455 分发饼干
孩子及其胃口：g[i]  
饼干及其尺寸：s[j]  
当分配给孩子的饼干尺寸 > 孩子的胃口 时孩子得到满足  
目标：让尽可能多数量的孩子得到满足，并输出这个最大数值  

## 思路
大尺寸的饼干可以让大小胃口的孩子都得到满足，但一个饼干只能分给一个孩子，所以大尺寸饼干应该尽可能给胃口大的孩子  

局部最优：充分使用当前饼干尺寸，满足胃口大的孩子  
全局最优：尽可能喂饱更多数量的孩子  

先把饼干尺寸和小孩胃口排升序  
从后向前遍历小孩数组，用大饼干优先满足胃口大的，同时统计满足小孩的数量  



PS：用大饼干先满足大胃口，外层for循环遍历的是胃口，然后取找满足胃口的饼干

时间复杂度：O(nlogn)
空间复杂度：O(1)


```python
# 大胃口优先

class solution:
    def findContentChildren(self, g, s):
        g.sort()    # 孩子胃口
        s.sort()    # 饼干尺寸
        index = len(s) - 1
        result = 0  # 统计满足孩子数量
        for i in range(len(g)-1, -1, -1): # 从最后一个孩子开始
            if index>=0 and s[index] >= g[i]:
                result += 1
                index -= 1
        return result
```
