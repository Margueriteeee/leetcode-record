```python
class Solution:
    def getIntersectionNode(self, headA:ListNode, headB:ListNode) -> ListNode:
        lenA, lenB = 0, 0
        cur = headA
        # A
        while cur:
            cur = cur.next
            lenA += 1
        
        cur = headB
        # B
        while cur:
            cur = cur.next
            lenB += 1

        curA, curB = headA, headB
        if lenA > lenB:  # curB, lenB
            curA, curB = curB, curA
            lenA, lenB = lenB, lenA
        
        for _ in range(lenB - lenA):  # curAcurB
            curB = curB.next
        while curA:
            if curA ==curB:
                return curA  #  curA  curB
            else:
                curA = curA.next
                curB = curB.next
        
        return None
```
