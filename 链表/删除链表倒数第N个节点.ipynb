```python
class ListNode:
    def __init__(self, val=0, next = None):
        self.val = val
        self.next = next

class Soluion:
    def removeElements(self, head:ListNode, n:int) -> ListNode:
        # 
        dummy_head = ListNode(0, head)
        low = fast = dummy_head

        for i in range(n + 1):  # fastn+1
            fast = fast.next
        
        while fast:
            low = low.next
            fast = fast.next
        
        # low
        low.next = low.next.next

        return dummy_head.next
```
