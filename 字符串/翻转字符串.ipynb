# 344 翻转字符串
输入： 字符串数组char[]  
要求：原地修改输入数组，仅有O(1)的额外空间   
输出：反转后的字符串数组  

### 直接可以用库函数解决的题目，不要用库函数！！！

## 思路：双指针法  
类似于反转链表，字符串也是一种数组，元素在内存中是连续分布的  
定义两个指针，从数组的头尾同时向中间移动，并交换元素  

```python
# 使用range
# n // 2 直接判断出要循环的次数，不用判断奇偶
# 奇数长度的字符串最中间不用交换  

class Solution:
    def reverseString(self, s:list[str]) -> None:
        n = len(s)
        for i in range(n // 2):
            s[i], s[n-i-1] = s[n-i-1], s[i]  # 首位对应位置交换即可
        
        # 切片翻转法：s[:] = s[::-1]
        # 列表推导法：s[:] [s[i] for i in range(n - 1, -1, -1)]  # 从最后一个索引开始赋值，依次递减
```
