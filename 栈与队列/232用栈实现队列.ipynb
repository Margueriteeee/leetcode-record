# 232 用栈实现队列
用栈的标准操作实现：  
push(x) -- 将一个元素放入队列的尾部。  
pop() -- 从队列首部移除元素。  
peek() -- 返回队列首部的元素。  
empty() -- 返回队列是否为空。  

## 思路
模拟类题目，用栈模拟队列的行为  
栈只能在数据的一端进行操作，要模拟队列必须要建立两个栈  
- 一个输入栈，一个输出栈  
- 二者开口方向相反  
- push操作直接把数据按顺序放入输入栈  
- pop操作则需要分情况：  
  - 输出栈为空：从输入栈一个个把数据拿出来放进输出栈，然后pop栈顶的元素  
  - 输出栈不为空：直接输出栈顶的元素即可  

peek函数和pop很像，功能上复用一下就可以了  

常用的功能和操作可以抽象成一个好用的函数或者工具类，**复用**的威力！！！
