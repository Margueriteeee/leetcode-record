# 1047 删除字符串中所有相邻的重复项
输入：小写字母组成的字符串  
要求：将所有相同且相邻的字母删除，删除后需要继续判断是否仍有重复  
输出：完成所有删除后的字符串，答案唯一  

## 思路
”相邻、相同” 本质上也是一种匹配题，栈来解决的经典题目  

类似于“消消乐”“对对碰”  

用栈来存放遍历过的元素遍历当前元素的时候查看栈中是否有相同数组的相邻元素  
- 有相同元素，同时相邻则删除  
- 遍历完字符串后，栈中剩下的元素依次出栈即可  
- 出栈后的元素还需要倒叙一下回复顺序  

## 补充
函数的递归实现就是利用栈的数据结构，每一次递归调用都会把函数的局部变量、参数值和返回地址压入调用栈中，递归返回的之后从栈顶弹出上一次递归的各项参数  

但如果无限递归，调用栈就可能会溢出，报错 Segementation fault  

```python
# 法一：使用栈
class Solution:
    def removeDuplicates(self, s:str)->str:
        res = list()
        for item in s:
            if res and res[-1] == item:  # 栈里有元素且栈顶元素和字符匹配则删除栈顶元素
                res.pop()
            else:
                res.append(item)   # 未匹配成功，存放访问过后的元素
        return "".join(res)   # 字符串拼接，这里直接按栈底-栈顶的顺序拼接了
```

```python
# 双指针模拟栈,需要把字符串转换为列表处理
# 不让用栈的备选
# fast指针遍历字符串的每个字符，读取新元素
# slow指针模拟栈的栈顶位置
# eg:abbbaa -> ab

class Solution:
    def removeDuplicates(self, s:str)->str:
        res = list(s)
        slow = fast = 0
        length = len(res)

        while fast < length:
            #  如果一样直接换，不一样会把后面的填在slow的位置
            res[slow] = res[fast]  # 把fast遍历到的新元素赋值到栈顶

            # 如果和前一个一样，指针后退一位
            if slow >0 and res[slow] == res[slow - 1]:
                slow -= 1   # 相当于栈顶元素出栈
            else:
                slow += 1
            fast += 1
        return ''.join(res[0: slow])  # 区间是左开右闭的，不包含slow

string = 'abbaa'
solution = Solution()
result = solution.removeDuplicates(string)
result
```
