# 栈和队列理论基础
## 定义
- 栈（stack）：先进后出，只能在固定的一端进行插入和删除操作    
- 队列（queue）：先进先出，队头出队，队尾入队

## 栈和队列的主要操作  
- push 和 pop
- 不提供走访功能，也不提供迭代器，不能通过iterator来遍历所有元素  
- 队列同样也不允许有遍历行为，不提供迭代器  

## python中栈和队列的底层实现
### 栈
动态数组或链表实现  

python中的**list**实际上就是动态数组，使用 **append()入栈 和 pop()出栈**  
操作只发生在列表的末尾，从而实现栈的定义  
- 插入和删除操作的速度非常快，时间复杂度O(1)  
- 用链表实现，操作栈顶也是O(1)的时间复杂度

### 队列
队列的操作发生在两端，python中通常使用 **collections.deque（双端队列）** 来实现  
实际上是一个双向链表  
- 在队头和队尾进行插入和删除操作，时间复杂度都是O(1)
