# 255 用队列实现栈
用队列的基本操作实现栈： 
push(x) -- 元素 x 入栈  
pop() -- 移除栈顶元素  
top() -- 获取栈顶元素  
empty() -- 返回栈是否为空  

## 思路
队列是队头出，队尾入 
相当于要在队列上实现单进单出  
两个按顺序进栈的数字1和2，按队列取数据的时候只能先取1，但栈的操作需要取出2  
这时候就需要把1存一下  

### 两个队列实现
一个队列负责进出的操作，另一个队列负责在出栈时把队列中先进队的元素存一下，并取出queue1队尾的元素实现出栈  

### 一个队列实现
队列实现栈的操作其实就是需要把最新入队（队尾）的那个元素取出  
用一个队列实现：  
- 把除了队尾元素的所有元素依次出队并加入队尾  
- 把队尾元素 “顶” 到队头  
- 直接出队即可
