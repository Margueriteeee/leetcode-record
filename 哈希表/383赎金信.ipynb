# 383 赎金信
你作为一个罪犯要写一封赎金信给家属，但是又不能暴露你的字迹  
你找到了一本杂志，从上面剪下你需要的字母组成赎金信中的单词，所以杂志上的每个字母只能使用一次  

输入：赎金信字符串ransom 和杂志字符串 magazine  
要求：判断ransom中需要的字符能不能有magazine中的字符构成，你可以假设两个字符串均只含有小写字母。  
输出：能 true 或 不能 false  

## 思路
典型的查找字符是否在另一个集合中出现的场景———— 哈希法  

注意：  
- magazine 中的字母不能重复使用  

- 只有小写字母，大小为26的哈希表就够了  

> 简直为数组量身定做的

- 用长度为 26 的数组记录magazine里所有字母出现的次数  

- 用ransom验证这个数组是否包含了所有其需要的字母  

> 这里的所有字母都是小写，就说明了数据量是小的，26长度的数组就够用了，用map反而更耗时  
> 时间复杂度: O(m+n)，其中m表示ransomNote的长度，n表示magazine的长度  
> 空间复杂度: O(1)  

```python
# 使用数组
# 布尔判断函数：all(iterable) 检查可迭代的对象中是否所有元素都为 真
class Solution:
    def canConstruct(self, ransomNote:str, magazine:str) -> bool:
        ransom_count = [0]* 26
        magazine_count = [0]* 26

        for i in ransomNote:
            ransom_count[ord(i) - ord('a')] += 1
        for i in magazine:
            magazine_count[ord(i) - ord('a')] += 1
        return all(ransom_count[i] <= magazine_count[i] for i in range(26))
```
