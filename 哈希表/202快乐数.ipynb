# 202 快乐数
> 写快乐数可以让你变得快乐 吧！

快乐数：  
- 正整数
- 每次迭代把每一位上的数字求平方和  
- 数次迭代后能够得到平方和为1 -> 快乐数
- 迭代无限循环则不是快乐数  

这个求和过程中，*“无限循环”* 说明：平方和sum可能会 **重复出现**   

想起来了吗？  
> 需要快速判断一个元素是否出现在集合中，使用哈希法！  
> 重复出现 -> 判断是否在一个集合中多次出现  

## 思路
用哈希表来判断每次迭代的平方和 sum 是否在表中重复出现  
sum = 1 返回true  

### 细节
取一个数的各个位上的单数：    
1. num % 10 取模（求余数）：取出最后一位上的数字  
2. num = num //10 整除（求商）：缩短位数继续迭代  

python中常用取商和余数的函数：  
divmod（a, b）：返回 a // b 和 a % b


```python
# 使用集合
# 往集合中添加数用 set.add(x)
# 往列表中添加数用 list.append(x)

class Solution:
    def isHappy(self, n:int) -> bool:
        record = set()

        while True:
            n = self.get_sum(n)
            if n == 1:
                return True
            
            # 如果结果重复出现则陷入死循环，不是快乐数
            if n in record:
                return False
            else:
                record.add(n)
    
    def get_sum(self, n:int) -> int:
        new_num = 0
        while n:
            n, r = divmod(n, 10)
            new_num += r**2
        return new_num
```

```python
# 使用数组
class Solution:
    def isHappy(self, n:int) -> bool:
        record = []
        while n not in record:
            record.append(n)
            new_num = 0
            n_str = str(n)     # 把数字转换成字符串取出每一位的值
            for i in n_str:
                new_num += int(i) ** 2   # 求平方和之前先转换成int
            if new_num == 1:
                return True
            else:
                n = new_num
        return False
```
