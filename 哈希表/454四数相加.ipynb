# 454 四数相加
输入：四个长度均为 N 的数组列表A,B,C,D，  
要求：计算有多少个元组 (i, j, k, l) ，使得 A[i] + B[j] + C[k] + D[l] = 0  
输出：元组(i, j, k, l)的个数  

## 思路
这道题其实是两数之和的double，把他拆解为两个两数之和就可以 

解题步骤：  
1. 定义一个映射，key：a，b两数之和；value：a，b两数之和出现的次数  

2. 遍历A、B两个数组，统计两个数组元素之和，以及出现的次数放到映射中  

3. 定义变量count，统计  a+b+c+d = 0 出现的次数

4. 遍历 C、D 数组，匹配 0 - (c + d) 在map中出现过的元素，存在就用count把map中出现的次数（value）统计出来  

5. 返回统计值count  

时间复杂度: O(n^2)  
空间复杂度: O(n^2)，最坏情况下A和B的值各不相同，相加产生的数字个数为 n^2  

```python
# 法1：使用字典
class Solution:
    def fourSumCount(self, nums1, nums2, nums3, nums4):
        # 初始化字典存储nums1和nums2中的元素之和及出现的次数
        hashmap = dict()
        for n1 in nums1:
            for n2 in nums2:
                # 下述判断也可以简化为：查找hashmap中n1+n2，存在则返回value，不存在返回默认的0
                # hashmap[n1 + n2] = hashmap.get(n1+n2, 0) + 1
                if n1+n2 in hashmap:
                    hashmap[n1 + n2] += 1   # 和已经出现过就增加1
                else:
                    hashmap[n1 + n2] = 1    # 否则初始化为1
        
        count = 0
        for n3 in nums3:
            for n4 in nums4:
                key = -n3 - n4
                if key in hashmap:
                    count += hashmap[key]  # 找到匹配值则把对应的次数统计到count
        return count
```

```python
# 使用带默认值的defaultdict
# defaultdict(lambda:0) 当键不存在时把返回值0作为该键的初始值
# 简洁版

from collections import defaultdict
class Solution:
    def fourSumCount(self, nums1, nums2, nums3, nums4):
        rec, cnt = defaultdict(lambda:0), 0
        for i in nums1:
            for j in nums2:
                rec[i+j] += 1
        
        for i in nums3:
            for j in nums4:
                cnt += rec.get(-i-j, 0)
        
        return cnt
```
