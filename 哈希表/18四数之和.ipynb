# 18 四数之和
输入：包含n个整数的数组nums和一个目标值target  
要求：判断nums中是否存在四个元素使得  a + b + c + d = target  
输出：所有满足条件且不重复的四元组

## 思路
这一题和三数之和是一个思路，用哈希法都会更加复杂一些

因此使用双指针发，在三数之和的基础上再嵌套一层for循环  
难在去重和剪枝

注意一个区别，三数之和中target是0，所以如果第一个数是0就直接返回空列表了  
但本题中target可以是任意数，这就意味着需要进一步的剪枝  
- 当前元素大于target，并且当前元素和target中任意一个大于零 的情况就可以直接排除

### 回忆双指针法的思路
- 一层for循环 nums[i] 为确定值  
- 循环内有left和right下标作为双指针
  - 找出三数之和=0的情况

### 四数之和
- 两层for循环 nums[i] 和 nums[j] 为确定值
- 循环内left和right两个指针下标
  - 找出四数之和 = target 的情况  
  - 

对于15.三数之和双指针法就是将原本暴力O(n^3)的解法，降为O(n^2)的解法  

四数之和的双指针解法就是将原本暴力O(n^4)的解法，降为O(n^3)的解法。

双指针法将时间复杂度：O(n^2)的解法优化为 O(n)的解法。也就是降一个数量级


```python
# 使用双指针
class Solution:
    def fourSum(self, nums:list[int], target:int) -> list[list[int]]:
        nums.sort()   # 升序排序
        n = len(nums)
        result = []

        for i in range(n):
            if nums[i] > target and nums[i] >0 and target > 0:  # 剪枝，可省
                break   # 终止整个循环
            if i > 0 and nums[i] == nums[i-1]:  # 去重
                continue  # 跳过当前轮次，继续下一个循环

            for j in range(i+1, n):
                if nums[i] + nums[j] > target and target > 0: # 剪枝，可省
                    break
                if j > i+1 and nums[j] == nums[j-1]:  # 去重
                    continue

                left, right = j+1, n-1

                while left < right:
                    s = nums[i] + nums[j] + nums[left] + nums[right]
                    if s == target:  # 和相等，先判断去重再移动双指针
                        result.append([nums[i], nums[j], nums[left], nums[right]])
                        while left < right and nums[left] == nums[left+1]:
                            left += 1   # 去重
                        while left < right and nums[right] == nums[right-1]:
                            right -= 1
                        left += 1
                        right -= 1
                    elif s < target:
                        left += 1  # 和偏小找更大的
                    else:
                        right -= 1  # 和偏大找更小的
            
            return result
```
