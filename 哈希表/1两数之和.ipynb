```python
# 
# enumerate(iterable, start=0) 
# enumerate index  value
class Solution:
    def twoSum(self, nums:list[int], target:int) -> list[int]:
        records = dict()

        for index, value in enumerate(nums):
            # keyvalue
            if target - value in records:  # key
                return [records[target - value], index]
            records[value] = index    # 
        return []
```
