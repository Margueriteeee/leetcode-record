# 15 三数之和  
输入：给定长度为n的整数数组nums 

要求：找到nums中是否存在三个数a,b,c，使得a+b+c=0

输出:所有满足条件且不重复的三元组列表

## 思路
按照求两数之和以及四数相加的思路，  
- 先用两层for循环确定两个数值  
- 再用哈希法查找第三个数 0-(a+b) 是否在数组中出现过  

思路是正确的，但怎么保证找到的三个数不重复呢？  

1. 把所有找到的三元组放进列表中，然后再遍历去重（费时）
   - 去重的过程中需要复杂的剪枝操作  
   - 面试中很难写出没有bug的代码  

2. 使用双指针法  

## 双指针法
比哈希法更高效，其实用了三个指针  

首先把数组由小到大排序

- 指针i：遍历数组，每次left和right指针相遇的时候移动  
- 指针left：定义在 i+1 的位置  
- 指针right：定义在数组末端  

每次循环中计算 nums[i] + nums[left] + nums[right]   

- 三指针之和 > 0, 说明元素大了，此时 right 指针向左移动找更小的数  
- 三指针之和 < 0, 说明元素小了，此时 left 指针向右移动找更大的数  

时间复杂度O(n^2)

PS：a, b, c三个元素去除重复的逻辑是，**遍历的时候遇到和上一个相同的元素就跳过**  

但在一个三元组中，abc三个元素是可以有重复的，只是三元组之间不能重复。这一点只需要在遍历的时候注意就可以避免   

```python
# 双指针法
# 推荐推荐！ 
class Solution:
    def threeSum(self, nums:list[int]) -> list[list[int]]:
        result = []
        nums.sort()   # 默认为升序

        for i in range(len(nums)):
            # 如果第一个元素已经 >0 ，已经是升序的数组中不可能再有和为0的三个数
            # 不需要进一步检查，直接返回空的result
            if nums[i] >0:
                return result

            # 跳过相同的元素以避免重复
            if i > 0 and nums[i] == nums[i - 1]:
                continue

            left = i + 1
            right = len(nums) - 1

            while right > left:
                sum_ = nums[i] + nums[left] + nums[right]

                if sum < 0:
                    left += 1
                elif sum > 0:
                    right -= 1
                else:
                    result.append([nums[i], nums[left], nums[right]])

                    # 跳过相同的元素避免重复
                    while right > left and nums[right] == nums[right - 1]:
                        right -= 1
                    while right > left and nums[left] == nums[left + 1]:
                        left += 1
                    
                    right -= 1
                    left += 1

        return result
```

```python
# 哈希法: 使用字典

class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        result = []
        nums.sort()

        # 找出a + b + c = 0
        # a = nums[i], b = nums[j], c = -(a + b)
        for i in range(len(nums)):
            if nums[i] > 0:
                break   # 排序后第一个元素大于0，不可能凑出和为0的三元组
            
            if i>0 and nums[i] == nums[i-1]:  #三元组元素a去重
                continue

            d = {}
            for j in range(i+1, len(nums)):
                if j > i + 2 and nums[j] == nums[j-1] == nums[j-2]: # 元素b去重
                    continue
                c = 0 - nums[i] - nums[j]
                if c in d:
                    result.append([i, nums[j], c])
                    d.pop(c)  # 三元组元素c去重
                else:
                    d[nums[j]] = j
        
        return result
```
